### Analysis Stats for Top 10

```python
top_10=df[:10]
```

```python
top_10.sort_values('A',ascending=False)
```

### Goal's vs Assist's for Top 10 Comparison

```python
x=list(top_10['Player'])
y=list(top_10['G'])
y1=list(top_10['A'])
plt.figure(figsize=(50, 30), dpi=100)
plt.bar(x,y,color ='maroon',width = 0.5,label='Goals')
plt.bar(x,y1,color='darkblue',width=0.5,label='Assists')
plt.xlabel("Player's Name")
plt.title("Goal's vs Assists for Top 10")
plt.legend()
plt.show()
```

### Goal's Vs Expected For Top 10

```python
x=list(top_10['Player'])
y=list(top_10['G'])
y1=list(top_10['xG'])
plt.figure(figsize=(50, 30), dpi=100)
plt.bar(x,y,color ='maroon',width = 0.5,label='Goals')
plt.bar(x,y1,color='darkblue',width=0.5,label='Expected Goals')
plt.xlabel("Player's Name")
plt.title("Goal's vs Expected Goals for Top 10")
plt.legend()
plt.show()
```

### Assist's vs Expected 

```python
x=list(top_10['Player'])
y=list(top_10['A'])
y1=list(top_10['xA'])
font = {'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
fig, axs = plt.subplots(2, sharex=True, sharey=True,figsize=(50,40))
fig.suptitle('Assits\'s vs Expected Assist\'s')
axs[0].bar(x, y)
axs[1].bar(x, y1)
```

```python
teams_goals = df.loc[df['G']>=15, "Team"]
plt.figure(figsize=(50, 30), dpi=100)
font = {'weight' : 'bold',
        'size'   : 22}
plt.rc('font', **font)
plt.hist(x=teams_goals,bins=5)
plt.xlabel("Team's Name")
plt.ylabel("Number Of Players")
plt.title("Teams with Player Scored More Than 15 Goal")
plt.show()
```

### Goal's for Top 10 and Their Apperence's

```python
x=list(top_10['Player'])
y=list(top_10['G'])
y0=list(top_10['G'])
y1=list(top_10['Apps'])
plt.figure(figsize=(50, 30), dpi=100)
plt.bar(x,y1,color ='maroon',width = 0.5,label='Goals')
plt.bar(x,y,color='darkblue',width=0.5,label='Apperence')
plt.xlabel("Player's Name")
plt.title("Goal's and Apperence's for Top 10")
plt.legend()
plt.show()
```
